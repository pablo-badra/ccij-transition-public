```python
def fix_cols(df):
    dic = {}
    for x in df.columns.values:
        print(x)
        val = input()
        dic[x] = val
    return dic

def shift_col(df):
    df = df
    val = df.columns.get_loc(df.columns[df.columns.str.contains('been deported', case=False)][0])
    dic = {}
    for x in range(val, df.shape[1]-2):
        dic[df.columns[x]] = df.columns[x+2]
    df.rename(mapper=dic, axis=1, inplace=True)
    return df

def to_datetime_3(df, crim_ent, other):
    frame = df
    frame = frame.convert_dtypes()
    print(frame.admin.values[:1])
    for x in other:
        try:
            frame[x] = pd.to_datetime(frame[x])
        except ValueError:
            print(x + " is messed up")
    for x in crim_ent:
        df[x].replace(to_replace=r'(?i).*unk.*', value=pd.NA, regex=True,inplace=True)
        try:
            frame[~frame[x].isna()][x] = pd.to_datetime(frame[~frame[x].isna()][x])
            frame[~frame[x].isna()][x] = pd.DatetimeIndex(frame[~frame[x].isna()][x]).year
        except ValueError:
            print(x + " is messed up")
    return frame

def non_digit(df, clean_digit):
    for x in clean_digit:
        df.loc[:, x].replace(to_replace=r'\D',value='',regex=True,inplace=True)
    return df

def consolidate(dfs, cols, drop, rename):
    for df in dfs:
        for x in cols:
            try:
                df[x].replace(to_replace='On', value='Yes', inplace=True)
                df[x].replace(to_replace='Off', value='No', inplace=True)
            except KeyError:
                pass
            except AttributeError:
                print(df + x + ' attributeerror')
        for y in drop:
            try:
                df.drop(labels=y, axis=1, inplace=True)
            except KeyError:
                pass
        df.rename(mapper=rename, axis=1, inplace=True, errors='ignore')
                
    return dfs

def func_1(dfs):
    vc = {}
    for df in dfs:
        for x in df.columns:
            vc[x] = vc.get(x, 0) + 1
    return vc

def count_col(df):
    c_n = []
    for x in dfs:
        for y in x.columns.values:
            c_n.append(y)
    return pd.Series(c_n)
```
